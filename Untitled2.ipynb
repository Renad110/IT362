1. Normalization using z-score

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

#Transform low, medium and high to numric values:
lvl_map = {'Low': 1, 'Medium': 2, 'High': 3}
df['Fatigue_Level'] = pd.to_numeric(df['Fatigue_Level'].replace(lvl_map), errors='coerce')

#Scale it:
df[['Fatigue_Level_z']] = scaler.fit_transform(df[['Fatigue_Level']])
print(df[['Fatigue_Level', 'Fatigue_Level_z']])
```

```python
# -------- Stress_Level --------
df['Stress_Level_num'] = pd.to_numeric(df['Stress_Level'].replace(lvl_map), errors='coerce')
df[['Stress_Level_z']] = scaler.fit_transform(df[['Stress_Level_num']])
print(df[['Stress_Level', 'Stress_Level_num', 'Stress_Level_z']])
```

```python
# -------- Crowd_Morale --------
#Transform Negative, Neutral and Positive to numric values:
lvl_map = {'Negative': 0, 'Neutral': 1, 'Positive': 2}
df['Crowd_Morale_num'] = pd.to_numeric(df['Crowd_Morale'].replace(lvl_map), errors='coerce')
df[['Crowd_Morale_z']] = scaler.fit_transform(df[['Crowd_Morale_num']])
print(df[['Crowd_Morale', 'Crowd_Morale_num', 'Crowd_Morale_z']])
```

```python
# -------- Interaction_Frequency --------
df[['Interaction_Frequency_z']] = scaler.fit_transform(df[['Interaction_Frequency']])
print(df[['Interaction_Frequency', 'Interaction_Frequency_z']].head(10))
```

## Statistical Summaries

### Subtask:
Generate descriptive statistics for both numerical and categorical columns to understand the central tendency, dispersion, and shape of the dataset's distribution.


**Reasoning**:
Generate descriptive statistics for numerical columns using the `.describe()` method and for categorical columns using `.describe(include='object')` to understand the central tendency, dispersion, and distribution of the dataset.



```python
print("\nDescriptive statistics for Numerical Columns:")
print(df.describe())

print("\nDescriptive statistics for Categorical Columns:")
print(df.describe(include='object'))
```

**Reasoning**:
Based on the previous step, the descriptive statistics for numerical and categorical columns have been generated. The next step is to visualize the distributions of these features to gain deeper insights. I will start by visualizing numerical features using histograms to understand their distribution, central tendency, and spread.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Plot histograms for numerical columns
plt.figure(figsize=(15, 2 * len(numerical_cols)))
for i, col in enumerate(numerical_cols):
    plt.subplot(len(numerical_cols) // 3 + 1, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

**Reasoning**:
After visualizing numerical distributions, the next logical step is to visualize categorical features using count plots to understand their distribution and frequency.



```python
categorical_cols = df.select_dtypes(include='object').columns

# Exclude 'Timestamp' and 'weather_dt' if they are still object type but represent dates/times and are not truly categorical for counting
categorical_cols = [col for col in categorical_cols if col not in ['Timestamp', 'weather_dt']]

# Determine the number of rows needed for subplots
num_plots = len(categorical_cols)
num_cols = 3
num_rows = (num_plots + num_cols - 1) // num_cols

plt.figure(figsize=(18, 5 * num_rows))
for i, col in enumerate(categorical_cols):
    plt.subplot(num_rows, num_cols, i + 1)
    sns.countplot(y=df[col], order=df[col].value_counts().index, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

**Reasoning**:
The previous code generated a FutureWarning regarding the use of 'palette' in `sns.countplot` without assigning 'hue'. To fix this and follow best practices, I will explicitly set `hue=df[col]` and `legend=False` in the `sns.countplot` call.



```python
categorical_cols = df.select_dtypes(include='object').columns

# Exclude 'Timestamp' and 'weather_dt' if they are still object type but represent dates/times and are not truly categorical for counting
categorical_cols = [col for col in categorical_cols if col not in ['Timestamp', 'weather_dt']]

# Determine the number of rows needed for subplots
num_plots = len(categorical_cols)
num_cols = 3
num_rows = (num_plots + num_cols - 1) // num_cols

plt.figure(figsize=(18, 5 * num_rows))
for i, col in enumerate(categorical_cols):
    plt.subplot(num_rows, num_cols, i + 1)
    sns.countplot(y=df[col], order=df[col].value_counts().index, palette='viridis', hue=df[col], legend=False)
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

## Correlation Analysis

### Subtask:
Compute the correlation matrix for all numerical columns and visualize it using a heatmap to identify relationships and dependencies between variables. Ensure the heatmap has a clear legend.


**Reasoning**:
To analyze the relationships between numerical features, I will compute the correlation matrix for all numerical columns and visualize it using a heatmap with annotations and a color bar for clarity.



```python
numerical_cols = df.select_dtypes(include=['number']).columns

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(18, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features', fontsize=16)
plt.show()
```
